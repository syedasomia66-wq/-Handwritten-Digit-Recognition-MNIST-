# <span style="color: #3A86FF; font-family: Georgia; font-size: 27px; text-shadow: 1px 1px 2px #A2D2FF"> Understanding how everything fits in TensorFlow</span>

##### $TensorFlow$: Nodes :
A node represents a mathimatical operation (like addition, multiplication, matrix multiplication, etc.).
##### $TensorFlow$: Placeholder :
A placeholder is a special tensor you define in the computation graph but don’t assign a value to right away.
Instead, you feed data later when running the graph. Commonly used for inputs (training data, test data, etc.) in TensorFlow 1.x.
##### $TensorFlow$: constant :
A constant is a fixed value tensor that never changes during execution. Useful for values like learning rates, fixed matrices, etc.
##### $TensorFlow$: variables :
A variable is a tensor whose value can change during execution. Typically used for weights and biases in machine learning models.
##### $TensorFlow$: session :
A session is an environment in TensorFlow 1.x where the computation graph is executed.You first build the graph (placeholders, constants, variables, operations),then run it inside a session.
##### $TensorFlow$: feed_dictionary :
feed_dict is a dictionary that supplies actual values for placeholders when running a TensorFlow graph.

# <span style="color: #3A86FF; font-family: Georgia; font-size: 27px; text-shadow: 1px 1px 2px #A2D2FF">Set Up TensorFlow Graph</span>

```python
tf.disable_v2_behavior()

X = tf.placeholder(tf.float32, shape=[None, TOTAL_INPUTS], name='X')
Y = tf.placeholder(tf.float32, shape=[None, NR_CLASSES], name='labels')
```

# <span style="color: #3A86FF; font-family: Georgia; font-size: 22px; text-shadow: 1px 1px 2px #A2D2FF">Hyperparameters</span>


```python
nr_epochs = 50
learning_rate = 1e-3

n_hidden1 = 512
n_hidden2 = 64
```

# [ 1 ]

# [ 2 ] 

# [ 3 ]

```python
def setup_layer(input, weight_dim, bias_dim, name):
    with tf.name_scope(name):
        initial_w = tf.truncated_normal(shape=weight_dim, stddev=0.1, seed=42)
        w = tf.Variable(initial_value=initial_w, name='W') 
        
        initial_b = tf.constant(value=0.0, shape=bias_dim)
        b = tf.Variable(initial_value=initial_b, name='B')
        
        layer_in = tf.matmul(input, w) + b
        
        if name == 'out':
            layer_out = tf.nn.softmax(layer_in)
        else:
            layer_out = tf.nn.relu(layer_in)
            
            tf.summary.histogram('weights',w)
            tf.summary.histogram('biases', b)
        
        return layer_out
```

```python
# Model with DropOut .

layer_1 = setup_layer(X, weight_dim=[TOTAL_INPUTS, n_hidden1],
                      bias_dim=[n_hidden1], name='layer_1')

Layer_drop = tf.nn.dropout(layer_1, keep_prob=0.8, name='dropout_layer')

layer_2 = setup_layer(Layer_drop, weight_dim=[n_hidden1, n_hidden2],
                      bias_dim=[n_hidden2], name='layer_2')

output = setup_layer(layer_2, weight_dim=[n_hidden2, NR_CLASSES],
                     bias_dim=[NR_CLASSES], name='out')

model_name = f'{n_hidden1}-DO-{n_hidden2} LR{learning_rate} E{nr_epochs}'
```

### Loss, Optimisation and Metrics

### Defining Loss Function

```python
with tf.name_scope('loss_calc'):
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=Y, logits=output))
```

### Defining Optimizer

```python
with tf.name_scope('optimizer'):
    optimizer = tf.train.AdamOptimizer(learning_rate)
    train_step = optimizer.minimize(loss)
```

### Accuracy Metric
any value that we have to visualis in tensorBoard is treced by the things callled summary
we created 2 different types of summary in this 1 one is called accuracy 2 one is call loss

```python
with tf.name_scope('accuracy_calc'):
    model_prediction = tf.argmax(output, axis=1, name='prediction')
    correct_pred = tf.equal(model_prediction , tf.argmax(Y, axis=1))
    accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

# <span style="color: #3A86FF; font-family: Georgia; font-size: 27px; text-shadow: 1px 1px 2px #A2D2FF">Batching The Data</span>

```python
size_of_batch = 1000
```

```python
num_examples = y_train.shape[0]
nr_iterations = int(num_examples/size_of_batch)

index_in_epoch = 0 
```

```python
# we gonna get all the values between 0 and 999 why bcz we gonna suopply 1000 to the size of the batch 
# so the starting value will be the 0 then index_in_epoch will be equal to itself + 1000 
# and the ending value will be equal to 1000.
# so in this will get value between 0:1000.
def next_batch(size_of_batch, data, labels):
    global num_examples
    global index_in_epoch

    start = index_in_epoch
    index_in_epoch += size_of_batch

    if index_in_epoch > num_examples:
        start = 0
        index_in_epoch = size_of_batch

    end = index_in_epoch
    return data[start:end], labels[start:end]
    
```

# <span style="color: #3A86FF; font-family: Georgia; font-size: 27px; text-shadow: 1px 1px 2px #A2D2FF">Training Loop</span>

```python
for epoch in range(nr_epochs):

    # =========== Training Dataset ===========
    for i in range(nr_iterations):
        batch_x, batch_y = next_batch(size_of_batch=size_of_batch, data=x_train, labels=y_train)
        feed_dictionary = {X: batch_x, Y: batch_y}  # Make sure Y is your labels placeholder
        sess.run(train_step, feed_dict=feed_dictionary)
        
        # Get summary and accuracy for the current batch
        s, batch_accuracy = sess.run(fetches=[merged_summary, accuracy], feed_dict=feed_dictionary)
        train_writer.add_summary(s, epoch * nr_iterations + i)
    
    # Calculate accuracy on the whole training set at the end of each epoch
    epoch_accuracy = sess.run(accuracy, feed_dict={X: x_train, Y: y_train})
    print(f'Epoch {epoch} \t Training Accuracy = {epoch_accuracy}')

    # =========== Validation Dataset ===========
    # Calculate accuracy on validation set at the end of each epoch
    summary, val_accuracy = sess.run(fetches=[merged_summary, accuracy], feed_dict={X: x_val, Y: y_val})
    validation_writer.add_summary(summary, epoch)
    print(f'Epoch {epoch} \t Validation Accuracy = {val_accuracy}')

print('Done training!')
```
